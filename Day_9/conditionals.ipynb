```python
# Level 1
# Driving age
age = int(input('Ingrese su edad: '))
if age >= 18:
    print('Tienes edad permitida para manejar')
elif age <= 0:
    print('Eres muy gracioso, denegado')
else:
    print('Necesitas', 18-age, 'años más para manejar')
```

```python
# Edades
my_age = 25
your_age = int(input('Ingrese su edad: '))
if my_age > your_age:
    print('Yo soy mayor que tu')
    if (my_age-your_age) == 1:
        print('Aunque solo por un año')
    else:
        print('Especificamente te llevo', my_age-your_age, 'años de diferencia')
elif my_age < your_age:
    print('Tu eres mayor que yo!')
    if (your_age-my_age) == 1:
        print('Aunque solo por un año')
    else:
        print('Especificamente me llevas', your_age-my_age, 'años de diferencia')
else:
    print('Tenemos la misma edad!, Somos de la misma quinta')
```

```python
a = int(input("Ingrese un primer numero: "))
b = int(input("Ingrese un segundo numero: "))
if a > b:
    print(a, 'es mayor que', b)
elif a == b:
    print(a, 'es igual a', b)
else:
    print(a, 'es menor que', b)
```

```python
# Level 2
'''
80-100, A
70-89, B
60-69, C
50-59, D
0-49, F
'''
print('Convertidor de puntuaje numerico a puntuaje alfabetico (A to F)\n')
score = int(input('Ingrese su puntuacion en el examen: '))
if score >= 80:
    print('Obtuviste una A')
elif score < 50: 
    print('Obtuviste una F') 
elif score >=50 and score < 60:
    print('Obtuviste una D')
elif score >=60 and score < 69:
    print('Obtuviste una C')
else:
    print('Obtuviste una B')
```

```python
# Temporadas
print('Determine en que temporada estamos!\n')
autumn = ('September', 'October', 'November')
winter = ('December', 'January', 'February')
spring = ('March', 'April', 'May')
summer = ('June', 'July', 'August')
mes = input('Ingrese el mes: ')
if mes in autumn:
    print('Estamos en Otoño')
elif mes in winter:
    print("Estamos en Invierno")
elif mes in spring:
    print("Estamos en Primavera")
elif mes in summer:
    print("Estamos en Verano")
else:
    print('No es un mes admitido, el formato es mes con inicial en mayuscula y en ingles')
```

```python
# Frutas fruits = ['banana', 'orange', 'mango', 'lemon']
fruits = ['banana', 'orange', 'mango', 'lemon']
print('Vamos a agregar mas frutas a la lista!\n')
fruta_nueva = input('Que fruta deseas agregar? ')
if fruta_nueva in fruits:
    print('Esa fruta ya esta en la lista, intenta con otra')
else:
    fruits.append(fruta_nueva)
    print('Claro, fruta agregada!')
    print(fruits)
```

```python
# Level 3, dado un diccionario
person={
    'first_name': 'Asabeneh',
    'last_name': 'Yetayeh',
    'age': 250,
    'country': 'Finland',
    'is_marred': True,
    'skills': ['JavaScript', 'React','Node', 'MongoDB', 'Python'],
    'address': {
        'street': 'Space street',
        'zipcode': '02210'
    }
    }
if 'skills' in person:
    print('Su habilidad del medio es:', person['skills'][len(person['skills'])//2])
    if 'Python' in person['skills']:
        print('Con que sabes usar Python')
    else:
        print('Veo que no sabes usar Python')

front_end = {'JavaScript', 'React'}
back_end = {'Node', 'MongoDB', 'Python'}
full_stack = front_end.union(back_end)
skills = set(person['skills'])
if front_end.issuperset(skills):
    print('Es un desarrollador Front End')
elif back_end.issuperset(skills):
    print('Es un desarrollador Back End')
elif full_stack.issubset(skills):
    print('Es un desarrollador Full Stack')
else:
    print('No clasificado')

if person['is_marred'] == True and person['country'] == 'Finland':
    print(person['first_name'], person['last_name'], 'vive en', person['country'], 'y esta casado')
```
